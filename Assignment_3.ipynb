```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files


from google.colab import files
uploaded = files.upload()

```

```python
data = pd.read_csv(io.BytesIO(uploaded['cleaned_student_dataset (2).csv']))
```

```python
print(data.shape)
print(data.columns)
data.head()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(x='gender', data=data)
plt.title('Gender Distribution of Students')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()



```

# ***Insight:***

---

This shows how many male and female students are in the dataset. If one gender dominates, it might indicate an imbalance in participation or sampling.

```python
plt.figure(figsize=(8,5))
sns.histplot(data['weekly_self_study_hours'], bins=10, kde=True)
plt.title('Distribution of Weekly Self Study Hours')
plt.xlabel('Weekly Self Study Hours')
plt.ylabel('Number of Students')
plt.show()
```

# **Insight:**

---


This visualizes how much time students spend studying weekly. Peaks may show the most common study duration, helping understand typical effort levels.

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='absence_days', y='math_score', data=data)
plt.title('Effect of Absence Days on Math Performance')
plt.xlabel('Absence Days')
plt.ylabel('Math Marks')
plt.show()
```

# **Insight:**

---


You can observe whether students who miss more days tend to score lower in math, highlighting attendance impact on performance.

```python
plt.figure(figsize=(8,5))
avg_scores = data.groupby('gender')[['math_score','history_score','physics_score','chemistry_score','biology_score','english_score','geography_score']].mean()
avg_scores.plot(kind='bar', figsize=(10,6))
plt.title('Average Subject Scores by Gender')
plt.ylabel('Average Marks')
plt.show()
```

# **Insight:**

---


This comparison reveals performance differences between genders across various subjects. It helps identify subjects where one group excels.

```python
plt.figure(figsize=(10,5))
sns.boxplot(x='career_aspiration', y='chemistry_score', data=data)
plt.title('Chemistry Marks by Career Aspiration')
plt.xticks(rotation=45)
plt.show()
```

# **Insight:**

---


This shows how students with different career goals perform in chemistry. For instance, aspiring engineers might score higher due to stronger interest in science subjects.
